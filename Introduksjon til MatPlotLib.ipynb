## En enkel graf

```python
import matplotlib.pyplot as plt # Importerer matplotlib med aliaset plt
import numpy as np # Importerer numpy med aliaset np
plt.style.use('fivethirtyeight') # Bruker den globale stilen "fivethirtyeight"
```

```python
# Lager x-verdier og y-verdier med numpy.
x_verdier = np.arange(-5, 5.1, 0.1)
y_verdier = x_verdier * x_verdier

# Plotter grafen med plot-metoden og show-metoden.
plt.plot(x_verdier, y_verdier)
plt.show()
```

## Flere grafer, stiler, og eksportering av bildet

```python
# Lager x-verdier og y-verdier
x_verdier = np.arange(-2, 2.1, 0.01)
y_verdier1 = x_verdier * x_verdier
y_verdier2 = x_verdier * x_verdier * x_verdier

# Plotter 3 grafer i samme figur med forskjellige farger, linjebredder, og linjestiler.
plt.plot(x_verdier, x_verdier, label="x opphøyd i første", color="#997766", linewidth=2)
plt.plot(x_verdier, y_verdier1, label="x opphøyd i annen", color="pink", linewidth=2)
plt.plot(x_verdier, y_verdier2, label="x opphøyd i tredje", linewidth=2, linestyle="dashed")

#Legger på tittel, xlabel, og ylabel på figuren.
plt.title("Polynomer")
plt.xlabel("tid (i sekunder)")
plt.ylabel("hastighet (i km/t)")

# Legger på en legend på figuren.
plt.legend()

# Lagrer bildet med filnavnet "Polynomer.png"
plt.savefig('Polynomer')

# Viser figuren her i Jupiter notebooks
plt.show()
```
