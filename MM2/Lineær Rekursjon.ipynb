## Lineær Rekursjon

I dette arket ser vi på hvordan vi kan bruke NumPy og MatPlotLib til å løse at lineær rekursjonsproblem.

```python
# Laster inn biblioteker og setter en stil.
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

Vi bruker NumPy til å bygge opp løsningsmatrisen.

```python
# Sette opp matrisen og vektoren
matrise = np.array([[1, 2], [-3.5, 4]])
vektor = np.array([[0], [1]])

# Bestemme antall omganger rekursjonen skal gjøre
antall_omganger = 5

# Definere løsning
løsning = vektor

# Generere nye verdier
for num in range(antall_omganger):
    løsning = np.hstack((løsning, matrise @ vektor))
    vektor = matrise @ vektor
    
# Skriv ut løsningen
print("Løsningen er: \n", løsning)

# Seperere løsningen i x-verdier og y-verdier
x_verdier = løsning[0, :]
y_verdier = løsning[1, :]
```

Vi kan nå bruke MatPlotLib til å plotte resultatene.

```python
# Verdier for linjene
plt.plot(x_verdier, linewidth=3, linestyle='--', alpha=0.7, label='x-verdier')
plt.plot(y_verdier, linewidth=3, linestyle='--', alpha=0.7, label='y-verdier')

# Design
plt.title('Arter i løpet av år')
plt.xlabel('Tid i år')
plt.ylabel('Antall individed')
plt.legend()

# Vise plottet
plt.show()
```
