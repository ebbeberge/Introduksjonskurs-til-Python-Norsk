## Enkel MatPlotLib

I dette arket bruker vi MatPlotLib til Ã¥ tegne grafer.

```python
# Importerer MatPlotLib, NumPy, og setter stilen til MatPlotLib
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
```

```python
# Setter verdier
x_verdier = np.arange(-10, 10, 0.01)
y_verdier = x_verdier ** 2

# Plotting
plt.plot(x_verdier, y_verdier, color="blue", linewidth=2, alpha=0.6, label="x squared")
plt.plot(x_verdier, y_verdier + 10, color="red", linewidth=2, alpha=0.6, label="x squared plus 10")

# Design
plt.title('Plotter $x ^ 2$')
plt.xlabel('x-verdier')
plt.ylabel('y-verdier')
plt.legend()

# Vise plottet
plt.show()
```

Under kan vi se hvordan vi kan plotte de to funksjonene i forskjellige vinduer.

```python
# Bruker subplots-funksjonen til Ã¥ lage en figur med akser.
fig, axs = plt.subplots(2)

# Plotter pÃ¥ hver av aksene individuelt
axs[0].plot(x_verdier, y_verdier, color="blue")
axs[1].plot(x_verdier, y_verdier + 10)
```
